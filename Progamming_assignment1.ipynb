```python
a = 60
b = 60
c = a * b
print("The number of seconds per hour is = ",c)
```

```python
seconds_per_hour = c
```

```python
daily_hours = 24
secs_d = daily_hours * seconds_per_hour
print("The number of seconds in a day is = ", secs_d)
```

```python
seconds_per_day = secs_d
```

```python
f = seconds_per_day / seconds_per_hour
print("The result from this division is = ", f)
```

```python
d = seconds_per_day // seconds_per_hour
print("The result from this division is = ", d)
```

```python
print(" In conclusion, the result from the last division agreed with the floating-point value from the previous division.")
```
